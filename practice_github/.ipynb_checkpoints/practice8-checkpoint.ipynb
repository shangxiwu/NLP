# Use Bi-gram model to create a word-level text generator 

```python
import nltk
import numpy as np
import random
import string

import bs4 as bs
import urllib.request
import re
from sklearn.feature_extraction.text import CountVectorizer

raw_html = urllib.request.urlopen('https://en.wikipedia.org/wiki/Tennis')
raw_html = raw_html.read()

article_html = bs.BeautifulSoup(raw_html, 'html.parser')
article_paragraphs = article_html.find_all('p')
article_text = ''

for para in article_paragraphs:
    article_text += para.text

article_text = article_text.lower()
print(article_text)
```

```python
article_text = re.sub(r'[^A-Za-z. ]', '', article_text)
```

```python
sentence_torkenize = article_text.split(".")
```

```python
ngrams = {}
words = 3

## write your code here
vect2 = CountVectorizer(ngram_range=(3, 3)).fit(sentence_torkenize)

print("Vocabulary:\n{}".format(vect2.get_feature_names()))
```
