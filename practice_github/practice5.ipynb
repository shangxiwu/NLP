# Read Speech file, use jieba to do word segmentation and do word frequency on each word

```python
import jieba
import json
from collections import Counter

ret = open("./dataset/speech.txt", "r").read().strip()
ret_ver2 = jieba.cut(ret, cut_all=False)
word_list = [word for word in ret_ver2]
# write your code here
# please reference count.csv
word_count = Counter(word_list)
print(word_count)

article = ''
for word in word_list:
    article = article + ' ' + word
article

```
